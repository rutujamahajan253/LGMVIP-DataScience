# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

# Import Dataset

```python
mnist= tf.keras.datasets.mnist
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
fig=plt.figure(figsize=(15,3))
for i in range(20):
  ax=fig.add_subplot(2,10,i+1)
  ax.imshow(np.squeeze(x_train[i]),cmap='PuBuGn')
  ax.set_title(y_train[i])
```

```python
print(x_train.shape)
print(x_train[0])
```

```python
xtrain = x_train/255.0
xtest = x_test/255.0
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape=(28,28)),
                                    tf.keras.layers.Dense(128, activation ='relu'),
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

```python
model.summary()
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam(),
              loss ='sparse_categorical_crossentropy',
            metrics=['accuracy'])
```

```python
model.fit(xtrain,y_train, epochs=5)
```

```python
model.fit(xtrain,y_train, epochs=9)
```

```python
print(model.evaluate(x_test,y_test))
```

```python
history=model.fit(x_train,y_train,epochs=10,batch_size=32,validation_split=0.1)
```

```python
plt.title("Accuracy and Loss")
plt.xlabel("Epoch")
plt.ylabel("acc/Loss")
plt.plot(history.history['accuracy'],label='acc')
plt.plot(history.history["val_accuracy"],label='val')
plt.show()
```

```python
plt.title("Accuracy and Loss")
plt.xlabel("Epoch")
plt.ylabel("acc/Loss")
plt.plot(history.history['loss'],label='acc')
plt.plot(history.history["val_loss"],label='val')
plt.show()
```

```python
plt.imshow(np.squeeze(x_test[1]),cmap="PuBuGn")
```

```python
prediction=model.predict(x_test)
print(np.argmax(prediction[0]))
```

```python
plt.imshow(np.squeeze(x_test[3]),cmap="PuBuGn")
```

```python
prediction=model.predict(x_test)
y_predicted_labels = [np.argmax(i)  for i in prediction]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt='d', cmap = 'coolwarm')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
plt.imshow(x_test[1],cmap='PuBuGn')
plt.title('Actual Value: {}'.format(y_test[1]))
prediction=model.predict(x_test)
plt.axis('off')
print('Predicted Value: ',np.argmax(prediction[1]))
if(y_test[1]==(np.argmax(prediction[1]))):
  print('Successful prediction')
else:
  print('Unsuccessful prediction')
```
