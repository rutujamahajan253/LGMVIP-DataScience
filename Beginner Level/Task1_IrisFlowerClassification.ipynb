# Splitting Dataset

```python
from sklearn.preprocessing import LabelEncoder
labelEncode = LabelEncoder()
```

```python
df['Flower Species'] = labelEncode.fit_transform = (df['Flower Species'])
df.head(10)
```

```python
x = df.drop(columns="Flower Species")
y = df["Flower Species"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

```python
print(x_train.head())
print(x_test.head())
print(y_train.head())
print(y_test.head())
```

```python
print("x_train: ", len(x_train))
print("x_test: ", len(x_test))
print("y_train: ", len(y_train))
print("y_test: ", len(y_test))
```

# Classification Models

```python
# Function for evaluation of model
def evaluate(model):
    pred=model.predict(x_test)
    print(classification_report(pred,y_test))
    confusion_matrix(model,y_test)
```

***Logistic Regression***

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, f1_score, recall_score, precision_score



def evaluate(model, X_train, X_test, y_train, y_test):
    y_test_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)

    print("\nTRAINING RESULTS: \n**********************************")
    clf_report = pd.DataFrame(classification_report(y_train, y_train_pred, output_dict=True))
    print(f"\nCONFUSION MATRIX:\n{confusion_matrix(y_train, y_train_pred)}")
    print(f"\nF1-SCORE:\n{f1_score(y_train, y_train_pred, average='macro')}")
    print(f"\nRECALL SCORE:\n{recall_score(y_train, y_train_pred, average='macro')}")
    print(f"\nPRECISION SCORE:\n{precision_score(y_train, y_train_pred, average='macro')}")
    print(f"\nACCURACY SCORE:\n{accuracy_score(y_train, y_train_pred):.4f}")
    print(f"\nCLASSIFICATION REPORT:\n{clf_report}")

    print("\nTESTING RESULTS: \n**********************************")
    clf_report = pd.DataFrame(classification_report(y_test, y_test_pred, output_dict=True))
    print(f"\nCONFUSION MATRIX:\n{confusion_matrix(y_test, y_test_pred)}")
    print(f"\nF1-SCORE:\n{f1_score(y_test, y_test_pred, average='macro')}")
    print(f"\nRECALL SCORE:\n{recall_score(y_test, y_test_pred, average='macro')}")
    print(f"\nPRECISION SCORE:\n{precision_score(y_test, y_test_pred, average='macro')}")
    print(f"\nACCURACY SCORE:\n{accuracy_score(y_test, y_test_pred):.4f}")
    print(f"\nCLASSIFICATION REPORT:\n{clf_report}")
```

```python
logReg = LogisticRegression(max_iter=1000)
logReg.fit(x_train, y_train)
print('\n\t\t\t*** Results for Logistic Regression ***')
evaluate(logReg, x_train, x_test, y_train, y_test)
```

***SVC***

```python
svc = SVC()
svc.fit(x_train, y_train)
print('\n\t\t\t*** Results for SVC ***')
evaluate(svc, x_train, x_test, y_train, y_test)
```

***Decision Tree Classifier***

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
print('\n\t\t\t*** Results for Decision Tree Classifier ***')
evaluate(dtc, x_train, x_test, y_train, y_test)
```

***KNN***

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
print('\n\t\t\t*** Results for KNN ***')
evaluate(knn, x_train, x_test, y_train, y_test)
```

Gaussian Naive Bayes

```python
gaussianNB = GaussianNB()
gaussianNB.fit(x_train, y_train)
print('\n\t\t\t*** Results for Gaussian Naive Bayes ***')
evaluate(gaussianNB, x_train, x_test, y_train, y_test)
```

***Multinomial Naive Bayes***

```python
multinomialNB = MultinomialNB()
multinomialNB.fit(x_train, y_train)
print('\n\t\t\t*** Results for Multinomial Naive Bayes ***')
evaluate(multinomialNB, x_train, x_test, y_train, y_test)
```

***MLP Classifier***

```python
mlp = MLPClassifier()
mlp.fit(x_train, y_train)
print('\n\t\t\t*** Results for MLP Classifier ***')
evaluate(mlp, x_train, x_test, y_train, y_test)
```
